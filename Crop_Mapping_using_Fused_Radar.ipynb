```python
import pandas as pd
import numpy as np
import pandas_profiling as pp
```

```python
# Read the file
data = pd.read_csv("/content/drive/MyDrive/WinnipegDataset.txt")
data
```

```python
# Exponential Data Analysis
data.describe()
```

```python
# Seperating data into test & train
X = data.iloc[:,1:]
y = data.iloc[:,:1]
```

```python
X.shape, y.shape
```

```python
X.head()
```

```python
y.head()
```

```python
# Converting to 1D array
y = np.ravel(y)
y
```

```python
# Finding Correlation among features
X.corr()
```

```python
# Create a correlation matrix
corr_matrix = X.corr().abs()
# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))
# Find the index of column where correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column]>0.95)]
# Drop Features
X.drop(X[to_drop],axis=1)
```

```python
# Preprocessing the data by StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.shape,y_train.shape,
```

```python
X_test.shape,y_test.shape
```

```python
X_train
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
# Importing Supprot Vector Machine
from sklearn import svm
svc = svm.SVC()
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
